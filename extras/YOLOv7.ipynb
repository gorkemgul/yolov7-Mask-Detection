## Making Predictions on Images using YOLOv7

```python
# Test the model on Custom Data
!python detect.py --weights yolov7.pt --conf-thres 0.5 --img-size 640 --source inference/images/people.jpg
```

```python
# Visualize the output of prediction
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg

# Define a function to visualize model's prediction
def visualizePred(Path):
  image = mpimg.imread(Path)
  plt.figure(figsize = (20, 20))
  plt.axis(False)
  plt.imshow(image);

# Use the function and see the result
visualizePred('runs/detect/exp2/people.jpg')
```
